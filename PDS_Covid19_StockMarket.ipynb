#  Step3: Use pandas to create a dataframe that aggregates and sums both confirmed cases and deaths on a global level

## Total number of confirmed cases aggregates and sums by country

```python
df_confirm = df_confirm1.groupby('Country/Region').sum().reset_index()
df_confirm
```

```python
df_confirm.info()
```

```python
df_confirm.isna().count()
```

## Total number of death cases aggregates and sums by country

```python
df_death = df_death1.groupby('Country/Region').sum().reset_index()
df_death
```

```python
df_death.info()
```

```python
df_death.isna().count()
```

## The number of Covid19 confirmed cases is the last day confirmed number

```python
df_covid_confirmed = df_confirm[['Country/Region','12/4/22']]
df_covid_confirmed.columns=['Country','Total_Confirmed']
df_covid_confirmed
```

## The number of COVID19 death cases is the last day death number

```python
df_covid_death = df_death[['Country/Region','12/4/22']]
df_covid_death.columns=['Country','Total_Death']
df_covid_death
```

## Merge confirmed cases dataframe and death cases data frame on a global level

```python
df_covid_global = pd.merge(df_covid_confirmed,df_covid_death,on='Country')
df_covid_global
```

The real estate market fluctuated at a low level and gradually rose from Apr. 2021 to Feb. 2022. It began a wave decline and is now at a three-year low.

Until Apr. 2022, the precious metals and real estate markets had moved in almost the opposite direction. Since then, it has fluctuated and is now low.
