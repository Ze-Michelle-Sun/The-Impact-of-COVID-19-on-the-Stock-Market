# Step2: Read the daily confirmed cases and deaths into two dataframes

```python
df_confirm1=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_confirm1
```

```python
df_death1=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
df_death1
```

#  Step3: Use pandas to create a dataframe that aggregates and sums both confirmed cases and deaths on a global level

## Total number of confirmed cases aggregates and sums by country

```python
df_confirm = df_confirm1.groupby('Country/Region').sum().reset_index()
df_confirm
```

```python
df_confirm.info()
```

```python
df_confirm.isna().count()
```

## Total number of death cases aggregates and sums by country

```python
df_death = df_death1.groupby('Country/Region').sum().reset_index()
df_death
```

```python
df_death.info()
```

```python
df_death.isna().count()
```

## The number of Covid19 confirmed is the last dayconfirmed number

```python
df_covid_confirmed = df_confirm[['Country/Region','11/27/22']]
df_covid_confirmed.columns=['Country','Total_Confirmed']
df_covid_confirmed
```

## The number of COVID19 death is the last day death number

```python
df_covid_death = df_death[['Country/Region','11/27/22']]
df_covid_death.columns=['Country','Total_Death']
df_covid_death
```

## Merge confirmed cases dataframe and deaths dataframe on a global level

```python
df_covid_global = pd.merge(df_covid_confirmed,df_covid_death,on='Country')
df_covid_global
```
