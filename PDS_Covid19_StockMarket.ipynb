```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from alpha_vantage.timeseries import TimeSeries
```

# Step2: Read the daily confirmed cases and deaths into two dataframes

```python
df_confirm1=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_confirm1
```

```python
df_death1=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
df_death1
```

#  Step3: Use pandas to create a dataframe that aggregates and sums both confirmed cases and deaths on a global level

## Total number of confirmed cases aggregates and sums by country

```python
df_confirm = df_confirm1.groupby('Country/Region').sum().reset_index()
df_confirm
```

```python
df_confirm.info()
```

```python
df_confirm.isna().count()
```

## Total number of death cases aggregates and sums by country

```python
df_death = df_death1.groupby('Country/Region').sum().reset_index()
df_death
```

```python
df_death.info()
```

```python
df_death.isna().count()
```

## The number of Covid19 confirmed is the last dayconfirmed number

```python
df_covid_confirmed = df_confirm[['Country/Region','11/27/22']]
df_covid_confirmed.columns=['Country','Total_Confirmed']
df_covid_confirmed
```

## The number of COVID19 death is the last day death number

```python
df_covid_death = df_death[['Country/Region','11/27/22']]
df_covid_death.columns=['Country','Total_Death']
df_covid_death
```

## Merge confirmed cases dataframe and deaths dataframe on a global level

```python
df_covid_global = pd.merge(df_covid_confirmed,df_covid_death,on='Country')
df_covid_global
```

## Create TimeSeries Object and Make API call

### Create 5 stock dataframes
   a) Overall American Market: American S&P 500 Index

   b) Overall Canadian Market: Canadian iShares S&P/TSX 60 Index Fund 
   
   c) Travel sector: Alimentation Couche-Tard (TSE:ATD.TO)
   
   d) The Real Estate sector: iShares Global Real Estate Index ETF
   
   e) Precious metals (Gold, Silver, Platinum, etc): iShares S&P/TSX Global Gold Index ETF

```python
ts =TimeSeries(key,output_format='pandas')
st_USA = ts.get_weekly_adjusted('SPY')
st_CAN = ts.get_weekly_adjusted('XIU.TO')
st_travel = ts.get_weekly_adjusted('ATD.TO')
st_real_estate = ts.get_weekly_adjusted('CGR.TO')
st_precious_metals = ts.get_weekly_adjusted('XGD.TO')
```

```python
df_USA = pd.DataFrame(st_USA[0])
df_CAN = pd.DataFrame(st_CAN[0])
df_travel = pd.DataFrame(st_travel[0])
df_real_estate = pd.DataFrame(st_real_estate[0])
df_precious_metals = pd.DataFrame(st_precious_metals[0])
```

## Add name column for each dataframe

```python
df_USA['Name']='America'
df_CAN['Name']='Canada'
df_travel['Name']='Travel'
df_real_estate['Name']='RealEstate'
df_precious_metals['Name']='PreciousMetals'
```

## Rename the name of high and low columns and drop other columns except date column
